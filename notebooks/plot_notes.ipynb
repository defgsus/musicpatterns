```python
def plot_pattern(*pattern, max_length=64):
    # print(notes)
    p = figure(plot_width=930, plot_height=250, y_axis_label="note", x_axis_label="step")
    # p.xaxis.formatter = get_datetime_formatter()
    for i, note in enumerate(pattern):
        color = colors.RGB(
            127 + 100 * math.sin(i/1.5),
            127 + 100 * math.sin(i/1.5*1.3),
            127 + 100 * math.cos(i/1.5*1.7),
        )
        y = list(MaxLength(note, max_length))
        y1 = list(MaxLength(note+1, max_length))
        x = list(range(len(y)))
        p.vbar(x, 0.9, y1, y, fill_color=color, line_color=colors.RGB(0,0,0))
        #p.scatter(x, y, radius=0.5, fill_color=color, line_color=None)
    show(p)
```

```python
n = EuclideanRhythm(4, 10)
#n = Range(10)
#n = Repeat(Range(5)) + Repeat(Range(7)) 
n = IsPrime(Range(offset=3)+1)
#n = FlatList(Divisors(Range()))
plot_pattern(n, max_length=64)
```

```python
notes = []
for a, b in ((2, 3), (3, 5), (5, 7), (7, 11)):
    notes.append(
        Repeat(Range(a)) + Repeat(Range(b))
    )         

plot_pattern(*notes, max_length=64)
```
